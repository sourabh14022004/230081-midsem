# Task 1.2 — Key Assumptions of CS4VM
**Paper:** Cost-Sensitive Semi-Supervised Support Vector Machine (CS4VM)


## Assumption 1

**Assumption:** The unlabeled data satisfy a **cluster assumption** — the decision boundary should pass through regions of low data density, and instances of the same class cluster together in the feature space (induced by kernel k).

**Why the method needs it:** The label mean estimation step (Eq. 7–8) maximises the margin *between the class-conditional means* m₊ and m₋. This only makes sense if the unlabeled data actually cluster by class: if they do not, the means are unreliable and the margin between them is meaningless. The entire S3VM lineage that CS4VM builds on is grounded in this assumption.

**Violation scenario:** A dataset where features are not predictive of class membership in any kernel space — for example, a checkerboard pattern or a dataset with heavy class overlap and noisy features — would violate this assumption. The estimated label means would mix both classes, producing biased estimates of m₊ and m₋.

**Paper reference:** Section "Related Work" (S3VM builds on the cluster assumption); Section "Estimating Label Means", Eq. (7)–(8); Figure 2 implicitly assumes well-separated class distributions.

---

## Assumption 2

**Assumption:** The **balance parameter r** (the expected number of positive unlabeled instances) is known or can be reliably estimated from the labeled data, i.e., u₊ = (r+u)/2 is set correctly by the user.

**Why the method needs it:** u₊ and u₋ determine how many unlabeled instances are assigned as positive and negative in the iterative label-mean estimation (Step 4). If r is wrong, the estimated label means m̂₊ and m̂₋ will be computed from the wrong number of instances, directly biasing the plug-in estimate in Eq. (6) and corrupting the final classifier. The balance constraint ŷ'1 = r in Eq. (2)–(3) is the mathematical embodiment of this assumption.

**Violation scenario:** A highly imbalanced dataset where the true class ratio in the unlabeled pool differs dramatically from the labeled sample ratio — for example, unlabeled medical records where 90% are negative but only 30% of the 10 labeled examples are negative — would lead to a severely misspecified r, undermining all downstream label-mean estimates.

**Paper reference:** Eq. (2) — balance constraint ŷ'1 = r; Section "CS4VM — Formulation": "u₊ and u₋ are obtained as the ratios of positive and negative examples in the labeled training data."

---

## Assumption 3

**Assumption:** The **misclassification costs are class-dependent** (not instance-dependent) and are fixed and known before training. Specifically, c(+1) and c(−1) are scalar constants that apply uniformly to all positive and negative instances respectively.

**Why the method needs it:** The cost values c(+1) and c(−1) are baked directly into the loss function ℓ(y, f(x)) = c(y)·max{0, 1−yf(x)} in Eq. (1)–(3), into the margin constraints in Eq. (7)–(8), and into the dual variable bounds in Eq. (9). The entire CS4VM derivation — including the asymmetric hinge loss in Figure 1 and the theoretical guarantee of Theorem 1 — depends on having a single fixed cost scalar per class. Instance-dependent costs would require a different formulation entirely.

**Violation scenario:** Fraud detection where each transaction has an individual monetary value (so misclassifying a 10,000-dollar fraud is far more costly than misclassifying a 10-dollar fraud) violates this assumption. In that scenario, the single scalar c(+1) cannot capture the heterogeneous per-instance cost, and the CS4VM objective would systematically underweight or overweight certain misclassifications.

**Paper reference:** Section "Related Work" — "the first one is class-dependent cost… the second one is example-dependent cost… the former occurs more often in real applications"; Eq. (1): ℓ(y, f(x)) = c(y)·max{0, 1−yf(x)}; Eq. (9): dual variable bounds 0 ≤ αᵢ ≤ c(yᵢ)C₁.
