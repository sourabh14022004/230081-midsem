# Task 1.3 — What the Paper Claims to Improve
**Paper:** Cost-Sensitive Semi-Supervised Support Vector Machine (CS4VM)


## Main Baseline

The primary baseline identified in the paper is **TSVM (Transductive SVM / Semi-Supervised SVM)** by Joachims (1999), extended naively to the cost-sensitive setting by incorporating misclassification cost for labeled examples only (as described in the Experiments section). The supervised **CS-SVM** (cost-sensitive SVM trained on labeled data alone) serves as the secondary supervised baseline, while Laplacian SVM is the other semi-supervised comparison.

---

## Limitation of the Baseline (TSVM + naive cost extension)

The paper identifies two entangled limitations:

1. **TSVM is cost-blind on unlabeled data.** TSVM uses a symmetric hinge loss for unlabeled instances (Figure 1, left panel). When c(+1) ≠ c(−1), simply incorporating costs for labeled data while ignoring the asymmetry for unlabeled data leads to suboptimal total cost — the method does not account for the fact that different types of errors on the unlabeled data carry different costs once their labels are revealed at test time.

2. **Non-continuous loss when costs differ.** As the paper notes in the "Label Means for CS4VM" section: when c(+1) ≠ c(−1), the loss function for unlabeled data becomes non-continuous, so standard optimization techniques used for TSVM (such as the concave-convex procedure by Collobert et al. 2006) cannot be directly applied. This makes it hard to extend TSVM cleanly to the cost-sensitive regime.

---

## How CS4VM Overcomes the Limitation

CS4VM avoids the non-continuous loss problem entirely by **reformulating the CS4VM objective in terms of label means** (Eq. 5–6), which decouples the cost weights from the discrete label signs. This converts the hard combinatorial problem into a convex QP (Eq. 9) solvable by standard SMO, while naturally incorporating cost asymmetry into both the labeled and unlabeled data terms in the objective.

---

## One Condition Where CS4VM Would Not Outperform the Baseline

**When the cluster assumption fails severely and labeled data is relatively informative.** CS4VM's advantage over supervised CS-SVM comes from the label means of the unlabeled data carrying useful distributional information. If the unlabeled data follow a distribution that is fundamentally different from the test distribution — for example, in a covariate shift scenario — the estimated label means m̂₊ and m̂₋ will be corrupted rather than informative. In such a case, the unlabeled data actively mislead the CS4VM, and a simple supervised CS-SVM (which ignores the unlabeled data entirely) would be the safer choice. The paper itself observes this on the Wdbc dataset (Table 1) where "the cluster assumption and manifold assumption do not hold," and CS4VM does not gain over the supervised CS-SVM for that dataset.
